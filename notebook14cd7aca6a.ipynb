```python
df=pd.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

**Text Preprocessing Steps**

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 10000):
  review = re.sub('[^a-zA-Z]', ' ', df['review'][i])
  review = re.sub('<.*?', ' ', df['review'][i])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(corpus).toarray()
y=df.iloc[:,-1].values[0:10000]
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
c1=GaussianNB()
c2=MultinomialNB()
c3=BernoulliNB()
c1.fit(x_train,y_train)
c2.fit(x_train,y_train)
c3.fit(x_train,y_train)
    
```

```python
y1=c1.predict(x_test)
y2=c2.predict(x_test)
y3=c3.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm1 = confusion_matrix(y_test, y1)
cm2 = confusion_matrix(y_test, y2)
cm3 = confusion_matrix(y_test, y3)
print(accuracy_score(y_test, y1))
print(accuracy_score(y_test, y2))
print(accuracy_score(y_test, y3))
```
